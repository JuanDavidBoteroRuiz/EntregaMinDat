```python
import pandas as pd 
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_movie = 'https://drive.google.com/uc?id=1H64TOo4OQYnDPn50x1zMkF0bZhqhYaeB'
url_rating = 'https://drive.google.com/uc?id=12vtDPA2TogM4fho6Av8hkcOQI0-BRYDd'
```

```python
peliculas = pd.read_csv(url_movie)
```

```python
# Se preparan los datos
peliculas['year'] = peliculas.title.str.extract('(\(\d\d\d\d\))',expand = False)
peliculas['year'] = peliculas.year.str.replace('(','').str.replace(')','') 
peliculas['title'] = peliculas.title.str.replace('(\(\d\d\d\d\))','')
peliculas['title'] =peliculas['title'].apply(lambda x: x.strip())
peliculas['genres'] = peliculas.genres.str.split('|') #Ahora se dividen los valores de la columna de generos en una lista de generos para simplificar el uso futuro, lo haremos con una funcion de cadena dividia de python
```

```python
rating = pd.read_csv(url_rating).drop('timestamp',1)
ratings = pd.merge(peliculas,rating).drop('year',1).drop('genres',1)
ratings
```

```python
movie_ratings = ratings.pivot_table(values="rating", index=["userId"], columns=["title"]) # Se genera una tabla con tantos registros como usuarios existan y con tantas columnas como peliculas existan, asignando la calificacion correspondiente.
movie_ratings.head()
```

```python
corr_matrix = movie_ratings.corr()
corr_matrix.head()
```

```python
corr_matrix.style.background_gradient(cmap='Greens')
```

```python
toystory_ratings = movie_ratings["Toy Story"] # nos traremos los rating de esta pelicula
toystory_ratings
```

```python
similar_movies = movie_ratings.corrwith(toystory_ratings) # se calcula la correlacion de dicha pelicula con todas las demas
similar_movies = similar_movies.dropna()
df = pd.DataFrame(similar_movies)
df.head(10)
```

```python
similar_movies.sort_values(ascending=False).head(10) # nos traemos las peliculas con mayor similaridad
```

```python
import numpy as np

movie_stats = ratings.groupby("title").rating.agg(["count","mean"]) # Se hace un contea y se calula la media de cada pelicula pasa saber cuales son las mÃ¡s calificadas y luego excluir las menos frecuentes
movie_stats.sort_values("count",ascending=False).head # se ordena segun el conteo
```

```python
threshold = 1000 #se asigna un limite inferior para filtrar las peliculas segun el conteo y sacar de la lista las menos calificadas
popular_movies = movie_stats['count'] >= threshold
movie_stats[popular_movies].sort_values(['mean'], ascending=False)[:20]
```

```python
df = movie_stats[popular_movies].join(pd.DataFrame(similar_movies, columns=['similarity'])) # asignamos la similaridad a cada pelicula
df.head()
```

```python
df.sort_values(['similarity'], ascending=False)[:20] # se ordenan las peliculas segun similaridad para que sean estas las peliculas a recomendar
```
