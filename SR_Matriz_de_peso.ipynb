```python
import pandas as pd 
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_movie = 'https://drive.google.com/uc?id=1H64TOo4OQYnDPn50x1zMkF0bZhqhYaeB'
url_rating = 'https://drive.google.com/uc?id=12vtDPA2TogM4fho6Av8hkcOQI0-BRYDd'
```

```python
peliculas = pd.read_csv(url_movie)
```

```python
rating = pd.read_csv(url_rating)
```

```python
peliculas.head()
```

```python
# Se preparan los datos
peliculas['year'] = peliculas.title.str.extract('(\(\d\d\d\d\))',expand = False)
peliculas['year'] = peliculas.year.str.replace('(','').str.replace(')','') 
peliculas['title'] = peliculas.title.str.replace('(\(\d\d\d\d\))','')
peliculas['title'] =peliculas['title'].apply(lambda x: x.strip())
peliculas['genres'] = peliculas.genres.str.split('|') #Ahora se dividen los valores de la columna de generos en una lista de generos para simplificar el uso futuro, lo haremos con una funcion de cadena dividia de python
```

```python
peliculas.head()
```

```python
#Ya con los generos en una lista se crean columnas y uilizando la tecnica one hot encoding para codificar las peliculas
peliculas_co = peliculas.copy() # se hace una copia
for index, row in peliculas.iterrows(): 
    for genre in row ['genres']: #se asigna 1 a cada genero que correspnda 
        peliculas_co.at[index,genre] =1 #se guarda en la copia
peliculas_co = peliculas_co.fillna(0) #se asigna 0 donde no aplican los generos
peliculas_co
```

```python
#Identificamos de forma visual y numérica cuales son las correlaciones mas fuertes entre los generos
peliculas_co.corr().style.background_gradient(cmap='Greens')
```

```python
rating = pd.read_csv(url_rating)
rating = rating.drop('timestamp',1)
```

```python
#Se crea un dataframe con voles de entrada de un usuario 
usuario_en =[
# {'title':'Breakfast Club, The','rating':5},
{'title':'Toy Story','rating':3.5},
# {'title':'Jumanji','rating':2},
# {'title':'Pulp Fiction','rating':5},
# {'title':'Akira','rating':4.5},
]
entrada_peli = pd.DataFrame(usuario_en)
entrada_peli
```

```python
# ahora se agrega la identificacion a cada pelicula que ingreso el usuario. Se hace filtrando las filas que contienen el titulo de la pelicula y luego fusionando con nuestra matriz de peliculas originales
id= peliculas[peliculas['title'].isin(entrada_peli['title'].tolist())] # se filtra
entrada_peli = pd.merge(id,entrada_peli)# se fusionan las matrices 
entrada_peli = entrada_peli.drop('genres',1).drop('year',1) 
entrada_peli
```

```python
# Se crea un nuevo dataframe solo con las peliculas del usuario y los generos
peli_usuario = peliculas_co[peliculas_co['movieId'].isin(entrada_peli['movieId'].tolist())]
peli_usuario 
```

```python
# Se crea un data frame con todos los generos pero solamente para las peliculas del usuario
peli_usuario = peli_usuario.reset_index(drop = True)
tabla_generos = peli_usuario.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
tabla_generos
```

```python
# Para modelar el algoritmo se crea la matriz de peso que es un producto punto entre la matriz transpuesta de los valores de los generos y los rating de entrada del usuario
perfil_usu= tabla_generos.transpose().dot(entrada_peli['rating'])
perfil_usu
```

```python
#ahora se a extraen los generos de la tabala original, para luego ser multiplicado por la matriz de peso
generos = peliculas_co.set_index(peliculas_co['movieId'])
generos = generos.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
generos
```

```python
#ahora se calcula el promedio ponderado para recomendar las peliculas deseadas de nuestra base de datos
recom = ((generos*perfil_usu).sum(axis=1))/(perfil_usu.sum())
recom = recom.sort_values(ascending = False) #ordenamos las recomendacion de forma descendente
recom.head(20)
```
